```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df_copy1  = df
```

```python
df_copy1.isnull().sum()
```

```python
df_copy1.shape
```

```python
df_copy1.columns = df_copy1.columns.str.strip()
```

```python
df_copy1.head()
```

```python
df_copy = df_copy1
```

```python
df_copy.head()
```

```python
df_copy['Embarked'] = df_copy['Embarked'].fillna(df_copy['Embarked'].mode()[0])
```

```python
df_copy['Sex'].value_counts()
```

```python
df_copy['Age'] = df_copy['Age'].fillna(df_copy.groupby('Sex')['Age'].transform('mean'))
```

```python
df_copy.isnull().sum()
```

```python
df_copy.drop(['PassengerId','Ticket','Name'],axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
df_copy['Embarked'].value_counts()
```

```python
df_copy = pd.get_dummies(df_copy, columns=['Embarked'], prefix='Embarked')
```

```python
df_copy.head()
```

```python
df_copy['Embarked_C'] = df_copy['Embarked_C'].apply(lambda x:1 if x==True else 0)
df_copy['Embarked_Q'] = df_copy['Embarked_Q'].apply(lambda x:1 if x==True else 0)
df_copy['Embarked_S'] = df_copy['Embarked_S'].apply(lambda x:1 if x==True else 0)
```

```python
df_copy.head()
```

```python
df_copy['Sex'] = df_copy['Sex'].apply(lambda x:1 if x=='male' else 0)
```

```python
df_copy.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_copy.drop('Survived',axis=1)
y = df_copy['Survived']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
score = accuracy_score(y_test,prediction)
print(score)
cm = confusion_matrix(y_test,prediction)
print(cm)
```
